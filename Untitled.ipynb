```python
# funxtion to frop the feature which have zero std, which have zero variability, 
```

```python
def get_cols_zero_std(df :pd.DataFrame):
    cols_to_drop=[]
    num_col=[i for i in df.columns if df[i].dtype !='O']
    for i in num_col:
        if df[i].std()==0:
            cols_to_drop.append(i)
    return cols_to_drop
    
```

```python
def get_reduntant_col(df:pd.DataFrame,missing_thresh=0.7):
    cols_missing_ratio=df.isnull().sum().div(df.shape[0])
    cols_to_drop= list(cols_missing_ratio[cols_missing_ratio>missing_thresh].index)
    return cols_to_drop
```

```python
cols_drop_1=get_cols_zero_std(df=df_train)
```

```python
cols_drop_1
```

```python
cols_drop_2=get_reduntant_col(df=df_train,missing_thresh=0.7)
```

```python
len(cols_drop_2)
```

```python
cols_to_drop=cols_drop_1+cols_drop_2+['Wafers']
```
